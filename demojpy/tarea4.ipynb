```python
import pandas as pd
import numpy as np
df = pd.read_csv('tlacuaches.csv')
df
```

```python
#Selecciono mis datos para el entrenamiento del sistema
train = df.iloc[:90,:]
train
```

```python
x_train = train.iloc[:,[5,6,7,8,9]]
x_train
```

```python
y_train = train.iloc[:,3]
y_train
```

```python
#Luego se agarro los datos sobrantes para realizar un seccion o conjunto de testeo
test = df.iloc[90:,:]
test
```

```python
x_test = test.iloc[:,[5,6,7,8,9]]
x_test
```

```python
y_test = test.iloc[:,3]
y_test
```

```python
#Se agrega la libreria de "Sklearn" 
from sklearn.svm import SVC
modelo = SVC(kernel='rbf', gamma=0.1, C=1)
modelo.fit(x_train, y_train)
```

```python
#En esta seccion, mando a realizar la precion con la seccion agarrada de testeo
predicted = modelo.predict(x_test)
predicted
```

```python
test = pd.DataFrame(test)
test['Prediccion del sexo del tlacuache'] = predicted
test
```

```python
#Seccion para visualizar la presicion de las predicciones
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
modeloNeigh = KNeighborsClassifier(n_neighbors=5)

#Yo escogui que solo se sacara la precision de 5 predicciones y el resultado es el mostrado en la impresion
scores = cross_val_score(modeloNeigh, x_train, y_train.values.ravel(), cv=5, scoring='accuracy')
print("Accuracy en cada interacion: ", scores)
```

```python
#Esta seccion es para mostrar la precision general o promedio
accuracy_promedio = np.mean(scores)
print("El accuracy promedio es de: ", accuracy_promedio)
```
