```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/OswaldoYair/searchMoreSalud/main/demojpy/Datasets%20Protecto%20mes%202/MY2022%20Fuel%20Consumption%20Ratings.csv')
```

```python
#Tratamiento de los valores nulos
df.isna().sum()
```

```python
#Atributo Model o modelo - Nueva clase
df['Model']=df['Model'].fillna('Desconocido')
df.isnull().sum()
```

```python
#Atributo Vehicle Class - Moda
df['Vehicle Class']=df['Vehicle Class'].fillna(df['Vehicle Class'].mode()[0])
df.isnull().sum()
```

```python
#Atributo Cylinders - Media
df['Cylinders']=df['Cylinders'].fillna(df['Cylinders'].median())
df.isnull().sum()
```

```python
#Atributo Transmission - Moda
df['Transmission']=df['Transmission'].fillna(df['Transmission'].mode()[0])
df.isnull().sum()
```

```python
#Atributo Fuel Consumption (City (L/100 km) y Fuel Consumption(Hwy (L/100 km))
df['Fuel Consumption (City (L/100 km)']=df['Fuel Consumption (City (L/100 km)'].fillna(df['Fuel Consumption (City (L/100 km)'].mean())
df['Fuel Consumption(Hwy (L/100 km))']=df['Fuel Consumption(Hwy (L/100 km))'].fillna(df['Fuel Consumption(Hwy (L/100 km))'].mean())
df.isnull().sum()
```

```python
#Codificacion Fuel type - One hot
df4 = pd.get_dummies(df, columns=['Fuel Type'], dtype=int)
df4.head()
```

```python
# Calcular la frecuencia de cada categoría de la clase Transmission
frecuencia = df4['Transmission'].value_counts(normalize=True).to_dict()
# Mapear la frecuencia de cada categoría al atributo
df4['Transmission'] = df4['Transmission'].map(frecuencia)

df4.head()
```

```python
# Calcular la frecuencia de cada categoría de la clase Vehicle class
frecuencia = df4['Vehicle Class'].value_counts(normalize=True).to_dict()
# Mapear la frecuencia de cada categoría al atributo
df4['Vehicle Class'] = df4['Vehicle Class'].map(frecuencia)

df4.head()
```

# **Esta seccion es una modificacion respecto a lo realizado anteriormente**



```python
#Codificacion del atributo de "Model Year","Make","Model" con OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
df4.iloc[:,[0,1,2]] = ordinal_encoder.fit_transform(df4.iloc[:,[0,1,2]])
df4.head(5)
```

```python
#Escalado de los atributos
from sklearn.preprocessing import MinMaxScaler
scalerMinMax = MinMaxScaler()
df4.iloc[:,[0,1,2,4,5,7,8,9,10,11,13]] = scalerMinMax.fit_transform(df4.iloc[:,[0,1,2,4,5,7,8,9,10,11,13]])
df4.head()
```

```python
#Seleccion de los mejores atributos para el entrenamiento de nuestros modelos
from sklearn.model_selection import train_test_split
x=df4.iloc[:,[0,1,2,4,5,7,8,9,10,11,13]]
y=df4.iloc[:,12]

#Utilizare la libreria de RandomForestClassifier
#Primero importamos la libreria y la funcion necesaria
from sklearn.ensemble import RandomForestClassifier

#Para dicho metodo, es necesario tener en cuenta que los valores de clasificacion deben ser numericos (0,1,2,3,etc)
#Como este data frame ya contiene este tipo de clasificacion, no es necesario realizar dicha conversion
#Sabiendo eso, creamos una variable con la funcionalidad del metodo
clf = RandomForestClassifier()

#Entrenamos el clasificador de bosque aleatorio
clf.fit(x,y)

# Obtiene la importancia de las características
importances = clf.feature_importances_

# Crea un DataFrame para visualizar las importancias
feature_importances = pd.DataFrame({'feature': x.columns, 'importance': importances})

# Ordena las características por importancia
sorted_features = feature_importances.sort_values(by='importance', ascending=False)

print(sorted_features)
```

```python
#Ahora seleccionaremos nuestros atributos y crearemos una nueva variable para almacenar los mas recomendables
x2 = x.iloc[:,[9,7,8,5,6,3,4]]
x2.info()

#x_train, x_test, y_train, y_test = train_test_split(x2,y,test_size=0.2,random_state=42)
```

# **Entrenamiento de los dos modelos para la clasificacion**

```python
#Empezamos con un modelo muy utlizado, el cual es el modelo de maquina de vectores, donde sacaremos un score, el cual es "Accuracy"
#Primero importamos dos librerias, la de SVC, y cross val score para el entrenamiento y calcular el resultado del score.
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

#Se crea un objeto con la funcion de la libreria Random Forest
smv = SVC()

#Se entrena el modelo con cross val score
scores = cross_val_score(smv, x2, y, scoring="accuracy", cv=5, n_jobs=-1)
print(scores)
print(scores.mean())
```

```python
#Ahora Utilizaremos un modelo de tipo "Red Neuronal", en este caso es para la clasificacion, el cual utilizare MLP(Red Neuronal Multicapa)

#Primero importamos la libreria
from sklearn.neural_network import MLPClassifier

#Ahora creamos un objeto con la funcion de la libreria importada y le agregaremos unos parametros para el metodo de entrenamiento
modeloMLP = MLPClassifier(hidden_layer_sizes = (16,40,12), max_iter=800, activation = 'relu', solver = 'adam')

#Al igual que en el anterior modelo, utilizamos la libreria k-fold cross validation para realizar el entrenamiento y sacar un score, que sigue siento "Accuracy"
scores = cross_val_score(modeloMLP, x2, y, scoring="accuracy", cv=5, n_jobs=-1)
print(scores)
print(scores.mean())
```

# **Ahora vamos con los dos modelos de regresion**

```python
rf = pd.read_csv('https://raw.githubusercontent.com/OswaldoYair/searchMoreSalud/main/demojpy/Datasets%20Protecto%20mes%202/CrabAgePrediction.csv')
rf.info()
```

```python
#Tratamiento de los valores nulos
rf.isna().sum()
```

```python
#Relleno datos nulos de la columna "Sex" con la moda
rf['Sex']=rf['Sex'].fillna(rf['Sex'].mode()[0])
rf.isnull().sum()
```

```python
#Los siguientes atributos por rellenar, son 7, los cuales utilizare la funcion de promedio para cada uno, y seria de la siguiente manera
rf.iloc[:,[1,2,3,4,5,6,7]]=rf.iloc[:,[1,2,3,4,5,6,7]].fillna(rf.iloc[:,[1,2,3,4,5,6,7]].mean())
rf.isnull().sum()
```

```python
#El ultimo atributo es la de "Age" y para este solo utlizare la mediana
rf['Age']=rf['Age'].fillna(rf['Age'].median())
rf.isnull().sum()
```

```python
#Codificacion "Sex" con "One hot"
rf = pd.get_dummies(rf, columns=['Sex'], dtype=int)
rf.isnull().sum()
```

```python
#Estandarizacion de datos
from sklearn.preprocessing import MinMaxScaler
scalerMinMax = MinMaxScaler()
rf.iloc[:,[0,1,2,4,5,6,7,8,9,10]] = scalerMinMax.fit_transform(rf.iloc[:,[0,1,2,4,5,6,7,8,9,10]])
rf.head()
```

```python
#Separacion de los atributos en X e Y
xr = rf.drop(['Weight'], axis=1)
yr = rf['Weight']
```

```python
xr.head(5)
```

```python
yr.head(5)
```

# **Entrenamiento de los dos modelos**

```python
#Comenzamos con un modelo de regresion muy utlizado, el cual es "RandomForestRegressor", donde sacaremos un score, el cual es r2.
#Primero importamos dos librerias, la de random forest, numpy para calcular el promedio y cross val score para el entrenamiento y calcular el resultado del score.
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import numpy as np

#Se crea un objeto con la funcion de la libreria Random Forest
modelo = RandomForestRegressor(random_state=None)

#Se entrena el modelo con cross val score
scoresClasifForest = cross_val_score(modelo, xr, yr, scoring="r2", cv=5, n_jobs=-1)
print(scoresClasifForest)
print(np.mean(scoresClasifForest))
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.metrics import r2_score

# Definir el modelo de red neuronal para regresión
#model = tf.keras.Sequential([
    #layers.Dense(units = 12, input_dim = 10, activation="relu"),
    #layers.Dense(64, activation='relu'),
    #layers.Dense(1)  # Capa de salida lineal para regresión
#])

# Compilar el modelo
#model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo
#model.fit(x_train, y_train, epochs=10, validation_data=(xr, yr))

#Para Metrica con cross validated
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold

def create_model():
  model = Sequential()
  model.add(Dense(12, input_dim=10, activation='relu'))
  model.add(Dense(64, activation='relu'))
  model.add(Dense(1, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model



kfold = KFold(n_splits=5, shuffle=True, random_state=42)

for train_index, test_index in kfold.split(xr):
    Xr_train, Xr_test = xr[train_index], xr[test_index]
    yr_train, yr_test = yr[train_index], yr[test_index]

    model = create_model()
    model.fit(Xr_train, yr_train, epochs=100, verbose=0)

    yr_pred = model.predict(Xr_test)
    yr_pred = tf.argmax(yr_pred, axis=1)
    yr_test = tf.constant(yr_test)

    accuracy = r2_score(yr_test, yr_pred)
    print(f'Accuracy: {accuracy}')

# Realizar predicciones con el modelo
#y_pred = model.predict(x_test)

# Calcular el coeficiente R2
#r2 = r2_score(y_test, y_pred)

#print("Coeficiente R2:", r2)
```
