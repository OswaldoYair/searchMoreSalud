```python
import pandas as pd
import numpy as np
df = pd.read_csv('Iris.csv')
df
```

```python
#Se selecciono un rango o conjunto para el entrenamiento que son de 100 datos de flores
train = df.iloc[30:,:]
train
```

```python
x_train = train.iloc[:,[1,3]]
x_train
```

```python
y_train = train.iloc[:,-1]
y_train
```

```python
#Luego se agarro los datos sobrantes para realizar un seccion o conjunto de testeo, que serian en total 50 datos
test = df.iloc[0:30,:]
test
```

```python
x_test = test.iloc[:,[1,3]]
x_test
```

```python
y_test = test.iloc[:,-1]
y_test
```

```python
#Se agrega la libreria de "Sklearn" 
from sklearn.svm import SVC
modelo = SVC(kernel='rbf', gamma=0.1, C=1)
modelo.fit(x_train, y_train)
```

```python
#En esta seccion, mando a realizar la precion con la seccion agarrada de testeo
predicted = modelo.predict(x_test)
predicted
```

```python
#Aqui se acomodan los resultados de prediccion pero visualizado en una tabla
test = pd.DataFrame(test)
test['Prediccion'] = predicted
test
```

```python
#Seccion para visualizar la presicion de las predicciones
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
modeloNeigh = KNeighborsClassifier(n_neighbors=5)

#Yo escogui que solo se sacara la precision de 5 predicciones y el resultado es el mostrado en la impresion
scores = cross_val_score(modeloNeigh, x_train, y_train.values.ravel(), cv=5, scoring='accuracy')
print("Accuracy en cada interacion: ", scores)
```

```python
#Esta seccion es para mostrar la precision general o promedio
accuracy_promedio = np.mean(scores)
print("El accuracy promedio es de: ", accuracy_promedio)
```
