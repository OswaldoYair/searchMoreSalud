```python
import pandas as pd
df = pd.read_csv('datasets/MY2022 Fuel Consumption Ratings.csv')
df['Model'].unique()
```

```python
df.head()
```

```python
df['Transmission'].unique()
```

```python
#Tratamiento de los valores nulos
df.isna().sum()
```

```python
#Atributo Model o modelo - Nueva clase
df['Model']=df['Model'].fillna('Desconocido')
df.isnull().sum()
```

```python
#Atributo Vehicle Class - Moda
df['Vehicle Class']=df['Vehicle Class'].fillna(df['Vehicle Class'].mode()[0])
df.isnull().sum()
```

```python
#Atributo Cylinders - Media
df['Cylinders']=df['Cylinders'].fillna(df['Cylinders'].median())
df.isnull().sum()
```

```python
#Atributo Transmission - Moda
df['Transmission']=df['Transmission'].fillna(df['Transmission'].mode()[0])
df.isnull().sum()
```

```python
#Atributo Fuel Consumption (City (L/100 km) y Fuel Consumption(Hwy (L/100 km))
df['Fuel Consumption (City (L/100 km)']=df['Fuel Consumption (City (L/100 km)'].fillna(df['Fuel Consumption (City (L/100 km)'].mean())
df['Fuel Consumption(Hwy (L/100 km))']=df['Fuel Consumption(Hwy (L/100 km))'].fillna(df['Fuel Consumption(Hwy (L/100 km))'].mean())
df.isnull().sum()
```

```python
#Codificacion Fuel type - One hot
df4 = pd.get_dummies(df, columns=['Fuel Type'], dtype=int)
df4.head()
```

```python
# Calcular la frecuencia de cada categoría de la clase Transmission
frecuencia = df4['Transmission'].value_counts(normalize=True).to_dict()
# Mapear la frecuencia de cada categoría al atributo
df4['Transmission'] = df4['Transmission'].map(frecuencia)

df4.head()
```

```python
# Calcular la frecuencia de cada categoría de la clase Vehicle class
frecuencia = df4['Vehicle Class'].value_counts(normalize=True).to_dict()
# Mapear la frecuencia de cada categoría al atributo
df4['Vehicle Class'] = df4['Vehicle Class'].map(frecuencia)

df4.head()
```

```python
#Escalado de los atributos
from sklearn.preprocessing import MinMaxScaler
scalerMinMax = MinMaxScaler()
df4.iloc[:,[4,5,7,8,9,10,11,13]] = scalerMinMax.fit_transform(df4.iloc[:,[4,5,7,8,9,10,11,13]])
df4.head()
```

```python
#Creacion de los dos modelos de aprendizaje empezando con el de maquina de vectores de soporte
from sklearn.model_selection import train_test_split
x=df4.iloc[:,[3,4,5,6,7,8,9,10,11,13]]
y=df4.iloc[:,12]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Aqui se puede visualizar tanto el momento donde se realiza la prediccion como el accuracy utilizando k-fold Cross Validation
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

#Se empieza a entrenar el modelo
smv = SVC()
smv.fit(x_train, y_train)

#Se realiza la prediccion
y_predic = smv.predict(x_test)

#Se saca el accuracy con cross validation
scores = cross_val_score(smv, x_test, y_test, cv=5, scoring='accuracy')
scores.mean()
```

```python
#Creacion del segundo modelo de aprendizaje KNN
from sklearn.neighbors import KNeighborsClassifier

k = 3 #Es el numero de neighbors
modelKNN = KNeighborsClassifier(n_neighbors=k)

#Se estrena el modelo
modelKNN.fit(x_train,y_train)

#Se realiza la prediccion
y_predic = modelKNN.predict(x_test)

#Se saca el accuracy con cross validation score
scores = cross_val_score(modelKNN, x_test, y_test, cv=5, scoring='accuracy')
scores.mean()
```
