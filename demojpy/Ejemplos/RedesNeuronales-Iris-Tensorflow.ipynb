```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/juarezefren/datasets/main/Iris.csv')
df.shape
df.head(5)
```

```python
df.head(5)
df['Species'].unique()
```

```python
df.info()
```

```python
df.drop(['Id'],axis=1, inplace=True)
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Species"] = le.fit_transform(df["Species"])
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:,:-1]
X.head(5)
```

```python
y = df.iloc[:,-1]
y.head(5)
```

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_x = scaler.fit_transform(X)
X = pd.DataFrame(scaled_x, columns=X.columns)
X.head(5)
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split as tts
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()

model.add(Dense(units = 12, input_dim = 4, activation="relu"))
model.add(Dense(units = 20, activation="relu"))
model.add(Dense(units = 3, activation="softmax"))
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

model.summary()
```

```python
acc=[]
for contador in range(0,5):
    x_train, x_test, y_train, y_test = tts(X,y, test_size=0.2)
    #print("Comenzando entrenamiento...")
    historial = model.fit(x_train,y_train, epochs = 300, verbose = False)
    #print("Modelo entrenado!")

    y_pred= model.predict(x_test)
    
    res= np.argmax(y_pred, axis=1)
    #res.shape
    r = accuracy_score(y_test,res)
    acc.append(r)
    
print(acc)
print('Promedio: ', np.mean(acc))
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de p√©rdida")
plt.plot(historial.history["loss"])
```

```python
from sklearn.metrics import confusion_matrix
C=confusion_matrix(res, y_test)
print(C)
```
