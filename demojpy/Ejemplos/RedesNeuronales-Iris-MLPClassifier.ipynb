```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/juarezefren/datasets/main/Iris.csv')
df.shape
```

```python
df.drop(['Id'],axis=1, inplace=True)
df.info()
```

```python
df.info()
```

```python
X = df.iloc[:,:-1]
X.head(5)
```

```python
y = df.iloc[:,-1]
y.head(5)
```

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_x = scaler.fit_transform(X)
X = pd.DataFrame(scaled_x, columns=X.columns)
X.head(5)
```

```python
X.describe()
```

```python
from sklearn.neural_network import MLPClassifier

modelo = MLPClassifier(hidden_layer_sizes = (16,40,12), max_iter=800, activation = 'relu', solver = 'adam')
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(modelo, X, y,
                         scoring="accuracy",
                         cv=5, n_jobs=-1)
print(scores)
print(scores.mean())
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = tts(X,y, test_size=0.2)
modelo.fit(x_train,y_train)
y_pred= modelo.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
C=confusion_matrix(y_pred, y_test)
print(C)

```
