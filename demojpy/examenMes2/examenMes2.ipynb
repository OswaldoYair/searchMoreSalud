```python
#Tratado de nulos
df['cat_dis']=df['cat_dis'].fillna(df['cat_dis'].mode()[0])
df.isnull().sum()
```

```python
df['crim']=df['crim'].fillna(df['crim'].mean())
df.isnull().sum()
```

```python
df['indus']=df['indus'].fillna(df['indus'].median())
df.isnull().sum()
```

```python
#Atributo Model o modelo - Nueva clase
df['pool']=df['pool'].fillna('Desconocido')
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=['cat_dis'], dtype=int)
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=['cat_age'], dtype=int)
df.isnull().sum()
```

```python
df = df.drop('pool', axis=1)
```

```python
df = df.drop('ID', axis=1)
```

```python
df.info()
```

```python
x = df.drop('medv', axis=1)
y = df['medv']
```

```python
x
```

```python
y
```

```python
#Ahora se realizara un ejemplo, pero utilizando arboles de decision, por lo que primero imcluimos la liberia
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()

regressor.fit(x, y)

importances = regressor.feature_importances_

feature_importances = pd.DataFrame({'Atributo': x.columns, 'Importancia': importances})

feature_importances = feature_importances.sort_values('Importancia', ascending=False)

feature_importances.head(17)
```

```python
xr = x.iloc[:,[11,0,12,6,3]]
xr
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

#Se crea un objeto con la funcion de la libreria Random Forest
modelo = RandomForestRegressor(random_state=None)

#Se entrena el modelo con cross val score
scoresClasifForest = cross_val_score(modelo, xr, y, scoring="r2", cv=5, n_jobs=-1)
print(scoresClasifForest)
print(np.mean(scoresClasifForest))
```
